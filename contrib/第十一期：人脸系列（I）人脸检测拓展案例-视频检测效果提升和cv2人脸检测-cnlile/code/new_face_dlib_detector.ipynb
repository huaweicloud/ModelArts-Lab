```python
import dlib
import numpy as np
import cv2
import time
from datetime import datetime
from IPython.display import clear_output, Image, display, HTML
```

```python
video_name = "./face_0.avi"
detector_kp = dlib.get_frontal_face_detector()
predictor_kp = dlib.shape_predictor("./models/shape_predictor_68_face_landmarks.dat")
```

```python
def keypoint_detector(image):
    global res_img
    img_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    dets = detector_kp(img_gray, 1)
    for k, d in enumerate(dets):    
        print("Detection {}: Left: {} Top: {} Right: {} Bottom: {}".format(
            k, d.left(), d.top(), d.right(), d.bottom()))
        # 计算矩形框大小
        height = d.bottom() - d.top()
        width = d.right() - d.left()
        print('height:',height, 'width:', width )
        res_img = cv2.rectangle(img_gray,( d.left(), d.top()), (d.right(), d.bottom()), 0, 1)
        shape = predictor_kp(img_gray, d)
        for i in range(68):
            res_img = cv2.circle(img_gray,(shape.part(i).x,shape.part(i).y), 1, 255, 2)
    return res_img
```

```python
def arrayShow(img):
    _,ret = cv2.imencode('.jpg', img) 
    return Image(data=ret) 

cap = cv2.VideoCapture(video_name)
start_time = datetime.now() #获得当前开始时间
count = 0
while True:
    try:
        clear_output(wait=True)
        ret, frame = cap.read()
        if ret:    
            res_img = keypoint_detector(frame)
            img = arrayShow(res_img)
            display(img)
            time.sleep(0.01)
            count = count + 1
        else:
            break
    except KeyboardInterrupt:
        cap.release()
cap.release()
end_time = datetime.now() #获得当前结束时间
durn = (end_time - start_time ).seconds  #两个时间差，并以秒显示出来
print('cost time:', durn, count)
```

```python
import sklearn.preprocessing as sp
import os
```

```python
# options用于设置训练的参数和模式
options = dlib.simple_object_detector_training_options()
options.add_left_right_image_flips = True
# 支持向量机的C参数，通常默认取为5.自己适当更改参数以达到最好的效果
options.C = 5
# 线程数，
options.num_threads = 8
options.be_verbose = True
```

```python
current_path = os.getcwd()
train_folder = current_path + '/dlib_face_detector_training_data/'
train_xml_path = train_folder + 'frontal_faces.xml'
print("training file path:" + train_xml_path)
```

```python
print("start training:")
dlib.train_simple_object_detector(train_xml_path, 'detector.svm', options)
```

```python
print("train finish.")
```
